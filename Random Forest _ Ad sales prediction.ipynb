```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

# Importing the DataSet

```python
dataset=pd.read_csv('Social_Network_Ads.csv') #load the dataset
X = dataset.iloc[:,:-1].values #independent variables
y = dataset.iloc[:,-1].values  #dependent variables
```

# Split train and test data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.25)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler  #feature scaling between the range -2 to 2
norm=StandardScaler()
X_train=norm.fit_transform(X_train)
X_test=norm.transform(X_test)
print(X_train)
```

# Model Training

```python
from sklearn.ensemble import RandomForestClassifier 
 # purity split and no of trees we want to build
classifier = RandomForestClassifier(n_estimators=10,criterion='entropy')
classifier.fit(X_train,y_train)
```

# Prediction

```python
pred_val=[30,87000] #the age and salary pred values
print(classifier.predict(norm.transform([pred_val]))) #prediction
```

```python
y_pred=classifier.predict(X_test)
print(y_pred) 
#comparing predict and test values of dependent variable
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
c_matrix=confusion_matrix(y_test,y_pred)  #confusion_matrix
print(c_matrix)
accuracy_score(y_test,y_pred)  #accuracy
```

# Train Data Visualisation

```python
from matplotlib.colors import ListedColormap
X_set, y_set = norm.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(norm.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Random Forest Classification (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

# Test Data Visualisation

```python
from matplotlib.colors import ListedColormap
X_set, y_set = norm.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(norm.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Random Forest Classification (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
